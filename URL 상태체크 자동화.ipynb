```python
import urllib.request
from urllib.request import urlopen
from urllib.error import URLError, HTTPError
import csv


list = open('./URL_LIST1.csv', 'r', encoding='UTF-8') #csv파일 open
url = csv.reader(list) 

#loop
for line in url:
    data = open('urlcheck.txt','a') #txt 파일 생성
    s = ",".join(line)

    try:
        res = urlopen(s)
        hap = s + ',' + str(res.status) #200 OK
        print(hap)
        print(hap,file=data)
    except HTTPError as e:             #HTTP error code
        #err = e.read()
        code = e.getcode()
        hap2 = s + ',' + str(code)
        print(hap2)
        print(hap2,file=data)
    except urllib.error.URLError as e: #URL error
        hap3 = s + ',' + str(e.reason)
        print(hap3)
        print(hap3,file=data)
    finally: 
        data.close()
```
